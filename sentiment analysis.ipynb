```python
#getting afinn dictionary in txt format and storing it as dictionary
d = {}
file=open("C:\\Users\\Saatwik Bisaria\\Downloads\\afinn.txt")
for line in file:
    ' '.join(line.split())
    x= line.split()
    d[x[0]] = x[1]
    
print(d)
```

```python
def checkKey(dict, key): 
      
    if key in dict: 
        return 1
    else: 
        return 0
```

```python
rating=0
```

```python
ratings={}
ratings=list()
tweets={}
tweets=list()
```

```python
#getting tweets for query demonitisation, excluding retweets , since the date of announcement , tweets in english language
for tweet in tweepy.Cursor(api.search, q='demonitisation' + ' -RT',since="2016-09-09",lang="en").items(2000):
    text=tweet.text.lower()
    for word in text.split():   #splitting tweets into words -tokenisation
        if(checkKey(d,word)==1):
                rating=rating+int(float(d[word]))
    print("\n",text)
    print("rating of tweet=",rating)
    ratings.append(rating)
    tweets.append(text)
    rating=0
```

```python
ratings=np.array(ratings)  #converting into numpy array
```

```python
#plotting graph
fig, ax = plt.subplots()
ax.plot(ratings)


# set the x-spine
ax.spines['left'].set_position('zero')

# turn off the right spine/ticks
ax.spines['right'].set_color('none')
ax.yaxis.tick_left()

# set the y-spine
ax.spines['bottom'].set_position('zero')

# turn off the top spine/ticks
ax.spines['top'].set_color('none')
ax.xaxis.tick_bottom()

plt.xlabel('#tweets')
plt.ylabel('ratings')
plt.title('twitter sentiment analysis')
plt.show()
```

```python
top_5_index = np.argsort(ratings)[-5:]
```

```python
print("TOP 5 POSITIVE TWEETS\n")
for i in range(0,len(top_5_index)):
    print(tweets[top_5_index[i]])
    print("tweet rating=",ratings[top_5_index[i]])
    print("\n")
```

```python
worst_5_index=ratings.argsort()[:5]
```

```python
print("TOP 5 NEGATIVE TWEETS\n")
for i in range(0,len(worst_5_index)):
    print(tweets[worst_5_index[i]])
    print("tweet rating=",ratings[worst_5_index[i]])
    print("\n")
```

```python
pos_count, neg_count, zero_count = 0, 0, 0
```

```python
for num in ratings: 
    if num == 0: 
        zero_count +=1
    elif num<0: 
        neg_count +=1
    else:
        pos_count +=1
```

```python
print("total number of tweets=",len(ratings))
print("total number of positive tweets=",pos_count)
print("total number of negative tweets=",neg_count)
print("total number of neutral",zero_count)
```

```python
pos=pos_count/len(ratings) * 100
neg=neg_count/len(ratings) * 100
zero=zero_count/len(ratings) * 100
print("percentage of positive tweets=",pos)
print("percentage of negative tweets=",neg)
print("percentage of neutral tweets=",zero)
```

```python
#plotting pie chart
labels = 'postive', 'negative', 'neutral', 
sizes = [pos_count, neg_count,zero_count]
explode = (0.1, 0, 0)  

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
